```python
def fast_object2cat(df):
    df_out = df.copy()
    object_cols = [df_out.dtypes.index[i] for i, col in enumerate(df_out.dtypes) if col == "object"]
    for col in object_cols:
        df_out[col] = df_out[col].astype('category')
    
    return df_out
```

```python
df_cat = fast_object2cat(df)
```

```python
def fast_cat2onehot(df):
    df_out = df.copy()
    cat_cols = [df_out.dtypes.index[i] for i, col in enumerate(df_out.dtypes) if col.name == "category"]
    df_out = pd.get_dummies(df_out, columns = cat_cols, drop_first=True)
    
    return df_out
```

```python
def barplot_cstm(x,y,data,xlabel,ylabel,figsize=(8,8),hue=None,palette=sns.color_palette("Oranges", 10),ci=95,n_samples=True):
    ''' ... '''
    fig, ax = plt.subplots()
    fig.set_size_inches(figsize)
    sns.barplot(x=x,y=y,data=data,hue=hue,palette=palette,ci=ci,ax=ax)
    ax.set(xlabel=xlabel, ylabel=ylabel)  

    nobs = data.groupby([x])[y].agg(['count'])
    nobs = ["n: " + str(i) for s in nobs.values for i in s]

    pos = range(len(nobs))
    if n_samples == True:
        for tick,label in zip(pos,ax.get_xticklabels()):
            ax.text(pos[tick], 0.03, nobs[tick], horizontalalignment='center', size='medium', color='k', weight='semibold')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    #ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    return ax
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
def hp_tuned_lr(X_train, y_train, X_test, y_test, gscv_scoring= "f1"):
    lr = LogisticRegression()

    # Regularization penalty
    penalty = ['l1', 'l2']
    # Regularization hyperparameter C
    C = np.linspace(0.01,1.5,50)
    # Hyperparameter space
    hyperparameters = dict(C=C, penalty=penalty)

    # Grid search using 5-fold cross validation
    lr_gscv = GridSearchCV(lr, hyperparameters, cv=5, verbose=2, n_jobs = 4, scoring = gscv_scoring)

    # Fit grid search
    lr_best = lr_gscv.fit(X_train, y_train)

    # Best hyperparameters
    print('Best Penalty:', lr_best.best_estimator_.get_params()['penalty'])
    print('Best C:', lr_best.best_estimator_.get_params()['C'])
    
    # Accuracy and F1-Score metrics
    print('Train Accuracy: ',accuracy_score(y_train,lr_best.predict(X_train)))
    print('Train F1-Score: ',f1_score(y_train,lr_best.predict(X_train)))
    print('Train ROC-AUC-Score: ',roc_auc_score(y_train,lr_best.predict_proba(X_train)[:,1]))

    # Accuracy and F1-Score metrics
    print('Test Accuracy: ',accuracy_score(y_test,lr_best.predict(X_test)))
    print('Test F1-Score: ',f1_score(y_test,lr_best.predict(X_test)))
    print('Test ROC-AUC-Score: ',roc_auc_score(y_test,lr_best.predict_proba(X_test)[:,1]))
    
    cm = confusion_matrix(y_test,lr_best.predict(X_test))
    plot_confusion_matrix(cm, ['Not accepted','Accepted'],
                              normalize=False,
                              title='Confusion matrix',
                              cmap=plt.cm.Oranges)

    return lr_best
```

```python
lr_raw_data = hp_tuned_lr(X_train, y_train, X_test, y_test, gscv_scoring= "f1")
```

```python
lr_std_data = hp_tuned_lr(X_train_std, y_train, X_test_std, y_test, gscv_scoring= "f1")
```

```python
def hp_tuned_svc(X_train, y_train, X_test, y_test, gscv_scoring= "f1"):
    svc = SVC(probability = True)

    # Penalty parameter C
    C = [0.1, 1, 10, 20, 30]
    # Kernel coefficient
    gamma = [0.0001, 0.0005, 0.0009]
    # Hyperparameter space
    hyperparameters = dict(C=C, gamma=gamma)

    # Grid search using 5-fold cross validation
    svc_gscv = GridSearchCV(svc, hyperparameters, cv=5, verbose=2, n_jobs = 4, scoring = gscv_scoring)

    # Fit grid search
    svc_best = svc_gscv.fit(X_train, y_train)

    # Best hyperparameters
    print('Best C:', svc_best.best_estimator_.get_params()['C'])
    print('Best gamma:', svc_best.best_estimator_.get_params()['gamma'])
    
    # Accuracy and F1-Score metrics
    print('Train Accuracy: ',accuracy_score(y_train,svc_best.predict(X_train)))
    print('Train F1-Score: ',f1_score(y_train,svc_best.predict(X_train)))
    print('Train ROC-AUC-Score: ',roc_auc_score(y_train,svc_best.predict_proba(X_train)[:,1]))

    # Accuracy and F1-Score metrics
    print('Test Accuracy: ',accuracy_score(y_test,svc_best.predict(X_test)))
    print('Test F1-Score: ',f1_score(y_test,svc_best.predict(X_test)))
    print('Test ROC-AUC-Score: ',roc_auc_score(y_test,svc_best.predict_proba(X_test)[:,1]))
    
    cm = confusion_matrix(y_test,svc_best.predict(X_test))
    plot_confusion_matrix(cm, ['Not accepted','Accepted'],
                              normalize=False,
                              title='Confusion matrix',
                              cmap=plt.cm.Oranges)

    return svc_best
```

```python
def hp_tuned_rf(X_train, y_train, X_test, y_test, gscv_scoring= "f1"):
    rf = RandomForestClassifier(random_state=23)
    
    # Criterion for quality of split
    criterion = ["gini","entropy"]
    # Number of trees in random forest
    n_estimators = [10,50,100]
    # Maximum number of levels in tree
    max_depth = [5,10,15]
    max_depth.append(None)
    # Minimum number of samples required to split a node
    min_samples_split = [5, 10, 20]
    # Minimum number of samples required at each leaf node
    min_samples_leaf = [1, 2, 4]
    # Method of selecting samples for training each tree
    bootstrap = [True, False]
    # Create hyperparameter space
    hyperparameters = dict(criterion=criterion,
                           n_estimators=n_estimators,
                           max_depth=max_depth,
                           min_samples_split=min_samples_split,
                           min_samples_leaf=min_samples_leaf,
                           bootstrap=bootstrap)

    # Grid search using 3-fold cross validation
    rf_gscv = GridSearchCV(rf, hyperparameters, cv=3, verbose=2, n_jobs = 4, scoring = gscv_scoring)

    # Fit grid search
    rf_best = rf_gscv.fit(X_train, y_train)

    # Best hyperparameters
    print('Best criterion:', rf_best.best_estimator_.get_params()['criterion'])
    print('Best n_estimators:', rf_best.best_estimator_.get_params()['n_estimators'])
    print('Best max_depth:', rf_best.best_estimator_.get_params()['max_depth'])
    print('Best min_samples_split:', rf_best.best_estimator_.get_params()['min_samples_split'])
    print('Best min_samples_leaf:', rf_best.best_estimator_.get_params()['min_samples_leaf'])
    print('Best bootstrap:', rf_best.best_estimator_.get_params()['bootstrap'])

    # Accuracy and F1-Score metrics
    print('Train Accuracy: ',accuracy_score(y_train,rf_best.predict(X_train)))
    print('Train F1-Score: ',f1_score(y_train,rf_best.predict(X_train)))
    print('Train ROC-AUC-Score: ',roc_auc_score(y_train,rf_best.predict_proba(X_train)[:,1]))
    
    # Accuracy and F1-Score metrics
    print('Test Accuracy: ',accuracy_score(y_test,rf_best.predict(X_test)))
    print('Test F1-Score: ',f1_score(y_test,rf_best.predict(X_test)))
    print('Test ROC-AUC-Score: ',roc_auc_score(y_test,rf_best.predict_proba(X_test)[:,1]))
    
    cm = confusion_matrix(y_test,rf_best.predict(X_test))
    plot_confusion_matrix(cm, ['Not accepted','Accepted'],
                              normalize=False,
                              title='Confusion matrix',
                              cmap=plt.cm.Oranges)

    return rf_best
```

```python
def hp_tuned_XGBC(X_train, y_train, X_test, y_test, gscv_scoring= "f1"):
    
    xgb = XGBClassifier(random_state=42)
    
    min_child_weight = [1, 5, 10]
    gamma = [0.5, 1, 1.5, 2, 5]
    n_estimators = [100,200,300]
    learning_rate = [0.1, 0.05, 0.01]
    max_depth = [4, 8]
    colsample_bytree = [0.6, 0.8, 1.0]
    subsample = [0.6, 0.8, 1.0]
    
    # Create hyperparameter space
    hyperparameters = dict(min_child_weight=min_child_weight,
                           gamma=gamma,
                           n_estimators = n_estimators,
                           learning_rate=learning_rate,
                           max_depth=max_depth,
                           colsample_bytree=colsample_bytree,
                           subsample=subsample)

    # Grid search using 3-fold cross validation
    XGBC_gscv = GridSearchCV(xgb, hyperparameters, cv=3, verbose=2, n_jobs = 4, scoring = gscv_scoring)

    # Fit grid search
    XGBC_best = XGBC_gscv.fit(X_train, y_train)

    # Best hyperparameters
    print('Best min_child_weight:', XGBC_best.best_estimator_.get_params()['min_child_weight'])
    print('Best gamma:', XGBC_best.best_estimator_.get_params()['gamma'])
    print('Best n_estimators:', XGBC_best.best_estimator_.get_params()['n_estimators'])
    print('Best learning_rate:', XGBC_best.best_estimator_.get_params()['learning_rate'])
    print('Best max_depth:', XGBC_best.best_estimator_.get_params()['max_depth'])
    print('Best colsample_bytree:', XGBC_best.best_estimator_.get_params()['colsample_bytree'])
    print('Best subsample:', XGBC_best.best_estimator_.get_params()['subsample'])

    # Accuracy and F1-Score metrics
    print('Train Accuracy: ',accuracy_score(y_train,XGBC_best.predict(X_train)))
    print('Train F1-Score: ',f1_score(y_train,XGBC_best.predict(X_train)))
    print('Train ROC-AUC-Score: ',roc_auc_score(y_train,XGBC_best.predict_proba(X_train)[:,1]))
    
    # Accuracy and F1-Score metrics
    print('Test Accuracy: ',accuracy_score(y_test,XGBC_best.predict(X_test)))
    print('Test F1-Score: ',f1_score(y_test,XGBC_best.predict(X_test)))
    print('Test ROC-AUC-Score: ',roc_auc_score(y_test,XGBC_best.predict_proba(X_test)[:,1]))
    
    cm = confusion_matrix(y_test,XGBC_best.predict(X_test))
    plot_confusion_matrix(cm, ['Not accepted','Accepted'],
                              normalize=False,
                              title='Confusion matrix',
                              cmap=plt.cm.Oranges)

    return XGBC_best
```

```python
import tensorflow as tf
from keras import backend as K

def auc(y_true, y_pred):
    auc = tf.metrics.auc(y_true, y_pred)[1]
    K.get_session().run(tf.local_variables_initializer())
    return auc
```

```python
def create_nn_clf(X_train, y_train, X_test, y_test):
    batch_size = 32
    num_classes = 2
    epochs = 100
    input_dim = X_train.shape[1]
    
    y_train_nn = keras.utils.to_categorical(y_train, num_classes)
    y_test_nn = keras.utils.to_categorical(y_test, num_classes)
    
    model = Sequential()
    model.add(Dense(20, input_dim=input_dim, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))

    early_stop = EarlyStopping(monitor='val_acc', patience = 4)

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=[auc])
    
    model.fit(X_train, y_train_nn,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, y_test_nn),
          callbacks=[early_stop])
    
    score = model.evaluate(X_test, y_test_nn, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    
    print('Accuracy: ',accuracy_score(y_test,model.predict_classes(X_test)))
    print('F1-Score: ',f1_score(y_test,model.predict_classes(X_test)))
    print('Test ROC-AUC-Score: ',roc_auc_score(y_test,model.predict_proba(X_test)[:,1]))
    cm = confusion_matrix(y_test,model.predict_classes(X_test))
    plot_confusion_matrix(cm, ['No Sale','Sale'],
                                normalize=False,
                                title='Confusion matrix',
                                cmap=plt.cm.Oranges)
    
    return model
```
